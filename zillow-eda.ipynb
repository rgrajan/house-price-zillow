```python
import pandas as pd

df = pd.read_csv('zecon/City_time_series.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.Date = pd.to_datetime(df.Date)
df2 = df[df['Date'] >='2016-1-1']
```

```python
df2.info()
```

```python
df.set_index('Date', inplace=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

weekly_mean = df[['ZHVI_1bedroom', 'ZHVI_2bedroom', 'ZHVI_3bedroom', 'ZHVI_4bedroom']].resample('M').mean()
print(weekly_mean.corr())
weekly_mean.plot(subplots=True, figsize=(20,10))
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(df3['ZHVI_1bedroom'])
plt.show()
```
